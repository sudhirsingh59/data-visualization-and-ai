## Why Matrix / Grid Plots?

Matrix or grid-style plots are perfect for:
- Showing correlations between variables  
- Visualizing pivot tables  
- Discovering patterns in large datasets


### Heatmaps (heatmap())

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")
sns.heatmap(tips.corr(numeric_only=True),annot=True)
```

### Cluster Map (clustermap())

```python
iris = sns.load_dataset("iris")
sns.clustermap(iris.corr(numeric_only=True),annot = True)
```

### Styling & Themes

```python
sns.set_style("darkgrid")
sns.set_palette("Set2")
sns.scatterplot(x = tips["total_bill"],y = tips['tip'],hue=tips['sex'],s=80)
```

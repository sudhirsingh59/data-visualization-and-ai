# Seaborn – Class 3: Distribution Plots

## What Are Distribution Plots?

Distribution plots show how data values are **spread or concentrated**.

Seaborn offers:
- **`histplot()`** → modern histogram (with optional KDE)
- **`kdeplot()`** → smooth curve of distribution
- **`rugplot()`** → small marks showing individual values  
*(Older `distplot()` is now replaced by `histplot()`.)*


## histplot() – Basic Histogram

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pylab as plt

tips = sns.load_dataset("tips")
tips.head()
```

```python
sns.histplot(tips['total_bill'],bins=50,kde=True,color="black")
plt.title("Histogram of Total Bill")
```

## kdeplot() – Kernel Density Estimate

```python
sns.kdeplot(tips['total_bill'],fill=True,color="green")
```

## distplot() (Legacy) vs histplot()

> `distplot()` is deprecated.
> Use `histplot()` or `kdeplot()` instead.


## rugplot() – Small Ticks for Observations

```python
sns.rugplot(tips['total_bill'],height=0.3,color='orange')
```

## Multiple Distributions with hue

```python
sns.histplot(x = tips['total_bill'],hue = tips['sex'],kde=True,bins=20,palette = "Set2")
```

Key takeaways

histplot() → flexible histogram (can combine with KDE).

kdeplot() → smooth density curve.

rugplot() → quick look at raw values.

distplot() is deprecated → prefer histplot().

Use hue to compare groups in the same distribution plot.
